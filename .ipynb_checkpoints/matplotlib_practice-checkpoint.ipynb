```python
import matplotlib.pyplot as plt 
```

```python
# Set the X Axis to a lit of strings for each month
x_axis= ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set Y-Axis to a list of floats as the total fare in US Dollars accummulated for each month 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
# MATPLOT INTERFACE

fig, ax= plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot 
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend 
plt.plot(x_axis,y_axis,marker="D", color="green", linewidth=2, label='Boston')

#plt.figure(figsize=(20,10))
#plt.subplots(figsize=(20,10))

# Create labels for the x and y axes 
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45 
plt.ylim(0,45)
#plt.xlim("Jan","June")
#plt.xlim(0,5)
# Create a title
plt.title("PyBer Fare by Month")

# Add a grid 
plt.grid()

# Add the Legend 
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Object Oriented way 

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create Vertical Bar Chart 
#Create the plot with ax.plot()

fig,ax= plt.subplots()
ax.bar(x_axis,y_axis, color= "black", label= "New York")
```

```python
# Create the Horizontal Bar chart  
fig, ax= plt.subplots()
ax.barh(x_axis,y_axis, color='cyan', label='Chicago')
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot 
#SKILL DRILL
plt.scatter(y_axis,x_axis, color='red', label='Boston')
plt.legend()
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title("PyBer Fare by Month")
plt.ylim('Dec','Jan')
plt.xlim(0,50)
```

```python
# Create a scatter plot 
plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
# CREATE WITH OBJECT-ORIENTED 
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20]
```

```python
# CREATE WITH OBJECT-ORIENTED 
# SKILL DRILL 
fig, ax= plt.subplots()
ax.scatter(x_axis,y_axis,s = [i*5 for i in y_axis], color="blue")
```

```python
# CREATE PIE CHART 

explode_values= (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode= explode_values, labels=x_axis, autopct= "%1.1f%%")
plt.show()
```

```python
# Assign 12 colors, one for each month 
colors= ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values=(0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
        autopct='%.1f%%',
       shadow=True)
plt.show()
```
